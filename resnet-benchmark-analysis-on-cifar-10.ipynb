## **Heading: ResNet Benchmark Analysis on CIFAR-10**

---

### **2. Quick Theory: CIFAR-10 & The ResNet Paradox**

* **Receptive Field:** CIFAR-10 images are only 32x32 pixels. In very deep models like **ResNet-152**, the network can easily "overthink" the image, leading to higher training times for marginal gains.
* **The "Accuracy Gap":** Notice that ResNet-18 and ResNet-101 are within **0.25%** of each other. This shows that for small datasets, a shallower model is often more efficient.
* **Optimization:** ResNet-152 achieved the highest accuracy (91.68%), but it required **2.5x more epochs** and **13x more time** than ResNet-18.

```python
import matplotlib.pyplot as plt

models = ['R-18', 'R-34', 'R-50', 'R-101', 'R-152']
acc = [87.92, 86.41, 87.93, 88.17, 91.68]
time_min = [23.6, 28.7, 95.4, 219.6, 312.1]

fig, ax1 = plt.subplots(figsize=(10, 5))

# Accuracy Bars
ax1.bar(models, acc, color='black', label='Accuracy %')
ax1.set_ylim(80, 100)
ax1.set_ylabel('Accuracy (%)')

# Runtime Line
ax2 = ax1.twinx()
ax2.plot(models, time_min, color='red', marker='o', label='Runtime (min)')
ax2.set_ylabel('Runtime (Minutes)')

plt.title('CIFAR-10: Accuracy vs. Training Cost (P100 GPU)')
plt.show()
```

### **Summary Conclusion**

For CIFAR-10, **ResNet-18** is the "sweet spot" for speed and performance. While **ResNet-152** is the most accurate, the "cost-per-percent" is extremely high, making it better suited for larger, more complex datasets like ImageNet.
